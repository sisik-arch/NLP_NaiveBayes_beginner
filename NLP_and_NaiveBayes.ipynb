```python
# convert into lower case
df['review_text'] = df['review_text'].str.lower()
df['review_text'].head()
```

```python
# remove punctuation marks : ?, !
#regualar expression
import re 

#create a fucntion called remove-apunctuation
def remove_punctuation(text):
    ##
    cleaned_text = re.sub(r"[^\w\s]", "", text)
    return cleaned_text
```

```python
#apply the remove_punctuation() fucntion to every row in the review_text column
df['review_text'] = df['review_text'].apply(remove_punctuation)

df['review_text'].head()
```

```python
## tokenization: use word_token
```

```python
import nltk

nltk.download('punkt')  # pretrained model usde by nltk to split tect into sentences or words
nltk.download('punkt_tab') # newer version  
```

```python
from nltk.tokenize import word_tokenize
```

```python
sample_review = df['review_text'][0]

tokens = word_tokenize(sample_review)

print(tokens)
```

```python
df['tokens'] = df['review_text'].apply(word_tokenize)

df[['review_text', 'tokens']].head()
```

```python
# STOP WORD REMOVAL 

# sentence -> tokenization -> check words in stop_words -> remove stop_words

# download required once only, NLT stores it locally
nltk.download('stopwords')

from nltk.corpus import stopwords


stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
#
def remove_stopwords(words):
    filtered_words = []

    for word in words:

        if word not in stop_words:

            filtered_words.append(word)

    return filtered_words

df["token_no_stopwords"] = df["tokens"].apply(remove_stopwords)

print(remove_stopwords)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
# convert reviews into vectors
X = tfidf.fit_transform(df['review_text'])
X
```

```python
# get all words (features) learned by the tfidf vectorizer
feature_names = tfidf.get_feature_names_out()

#print vocab
print(feature_names)
```

```python
import numpy as np 

#np.asarray() converts the sparse matrix result into a Numpy array 

avg_tfidf = np.asarray(X.mean(axis = 0)).ravel()

tfidf_scores = pd.DataFrame({
    'word' : feature_names,
    'avg_tfidf' : avg_tfidf   
})

tfidf_scores = tfidf_scores.sort_values('avg_tfidf', ascending = False)

print('Most Imp Words:')

print(tfidf_scores.head())
```

```python
print('Rare Words:')

print(tfidf_scores.tail(20))
```

```python
# sentiment analysis using ML 
#define features
X = tfidf.fit_transform(df['review_text'])

y = df['sentiment']
```

```python
# train-test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =0)
```

```python
# Train Naive Bayes

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)

print(accuracy)
```

```python
# classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
